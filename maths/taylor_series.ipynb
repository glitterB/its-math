## Taylor series

A function $f(x+\Delta x)$ may be taylor expanded about a base point $x$ (if $x$ is 'smooth' as x)

$f(x+\Delta x)=f(x)+\frac{df}{dx}\Delta x+\frac{d^{2}f}{dx^{2}}\frac{\Delta x^{2}}{2!}+..$

or

$f(x)$ could be expanded about point a:

$f(x+\Delta x)=f(a)+\frac{df}{dx}(a)(x-a)+\frac{d^{2}f}{dx^{2}}(a)\frac{(x-a)^{2}}{2!}+...$
$since x = a+ \Delta x$

## Taylor Series
In this Notebook, we will have a glance over Taylor Series.

```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial as fact
```

```python
#dummy data
x = np.linspace(-10,10,2000)
y = np.cos(x)
```

```python
#defing toylor series
def taylor(x,n):
    if n ==0:return 1
    term = ((-1)**n) * (x**(2*n)) / fact(2*n)
    return term + taylor(x, n - 1)
```

```python
# Plot the actual cosine function
plt.figure(figsize=(7, 4))
plt.plot(x, y, label='cos(x)', color='black', linewidth=2)
```

```python
# Plot Taylor series approximations from power 1 to 5
colors = ['red', 'blue', 'green', 'orange', 'purple']
for n in range(1, 6):
    y_taylor = [taylor(x, n) for x in x]
    plt.plot(x, y_taylor, label=f'Taylor n={n}', color=colors[n-1])
```

```python
# Graph details
plt.title('Cosine Function and Taylor Series Approximations')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
