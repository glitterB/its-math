```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fourier_square(x, n=15, A=1, L=np.pi):
    """
    Vectorized Fourier series approximation of a square wave.

    Parameters:
        x : array-like
            Input values.
        n : int
            Number of harmonics (use odd numbers only).
        A : float
            Amplitude of the square wave.
        L : float
            Half-period (so period = 2L).
    """
    # Odd harmonics
    k = np.arange(1, n + 1, 2)
    
    # Scale x for general period
    # sin(k * Ï€x / L)
    terms = np.sin(np.outer(k, np.pi * x / L)) / k[:, None]
    
    # Fourier sum
    y = (4 * A / np.pi) * np.sum(terms, axis=0)
    
    return y
```

```python
# Example usage
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = fourier_square(x, n=25, A=2, L=np.pi)

# Plotting

plt.plot(x, y)
plt.title("Fourier Series Approximation of a Square Wave")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
