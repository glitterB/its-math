Sobel Edge Detector is used to extact features of an image. This is used as filter in case of `Convolutional Neural Network`. Let's see SED in details.

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = plt.imread("C:/Users/rushi/Jupyter_Notebooks/se.jpg")
```

```python
#displaying image
plt.imshow(h,cmap='gray')
```

$
gx = 
\begin{pmatrix}
  -1       & -1     &-1\\
   0       &  0     & 0\\
   1       &  1     & 1\\
\end{pmatrix}
;
gy = 
\begin{pmatrix}
   1       &  1     & 1\\
   0       &  0     & 0\\
  -1       & -1     &-1\\
\end{pmatrix}
$

```python
gx = np.array([[-1,-1,-1],[0,0,0],[1,1,1]]).T
```

```python
gy = gx.T
gy[0,:] = gy[2,:]
gy[2,:]*=-1
```

```python
row = len(h)
col = len(h[0])

op1 = []
op2 = []
```

```python
for i in range(row-3):
    ar1 = []
    ar2 = []
    for j in range(col-3):
        ar1.append(np.sum(gx*h[i:i+3,j:j+3]))
        ar2.append(np.sum(gy*h[i:i+3,j:j+3]))
    op1.append(ar1)
    op2.append(ar2)
```

```python
op1 = np.array(op1)
op2 = np.array(op2)
plt.imshow(op1,cmap='gray')
```

```python
plt.imshow(op2,cmap='gray')
```

```python
plt.imshow(np.sqrt(np.square(op1)+np.square(op2)),cmap='gray')
```
